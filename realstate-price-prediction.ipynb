# 🏠 Real Estate Price Prediction — Regression Analysis

A comprehensive regression analysis project exploring how property features affect apartment prices using Python, pandas, and statsmodels.

## 📊 Project Overview

This analysis employs **Ordinary Least Squares (OLS) regression** to predict apartment prices based on key property characteristics including area, number of rooms, kitchens, bathrooms, and amenities.

### Objective
Identify and quantify the relationship between property features and selling prices to understand which factors have the most significant impact on real estate valuations.

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm;
import matplotlib.pyplot as plt
df=pd.read_csv('/kaggle/input/housing-prices-dataset/Housing.csv')
df.head()
```

## 1️⃣ Data Loading & Setup

Begin by importing necessary libraries and loading the housing dataset for analysis.

### Libraries Used
- **numpy**: Numerical computing
- **pandas**: Data manipulation and analysis
- **statsmodels**: Statistical modeling and OLS regression
- **matplotlib**: Data visualization

```python
neighborhood_dummies = pd.get_dummies(
    df[['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea','furnishingstatus']],
    dtype=int
)
df_new = df.join(neighborhood_dummies)
df_new.head()
```

## 2️⃣ Feature Engineering

Convert categorical variables into dummy variables to prepare data for regression analysis.

### Categorical Features Being Encoded
- Main road access
- Guest room availability
- Basement presence
- Hot water heating
- Air conditioning
- Preference area status
- Furnishing status

```python
df_new['intercept']=1
x=df_new[['intercept','area','bedrooms','bathrooms','stories','basement_yes','hotwaterheating_yes','airconditioning_yes','prefarea_yes','furnishingstatus_furnished','furnishingstatus_semi-furnished']]
lm = sm.OLS(df_new['price'], x)
results = lm.fit()
(results.summary())
```

## 3️⃣ Model Development & Fitting

Build and fit an **Ordinary Least Squares (OLS)** regression model to quantify the relationship between property features and prices.

### Model Specification
The linear regression model takes the form:

$$\text{Price} = \beta_0 + \beta_1(\text{Area}) + \beta_2(\text{Bedrooms}) + \beta_3(\text{Bathrooms}) + \ldots + \epsilon$$

Where coefficients represent the change in price for a one-unit increase in each feature.

```python
print(results.params.apply(lambda x: f"{x:,.2f}"))
```
